```python
import matplotlib.pyplot as plt
import numpy as np

def simulate_performance_graph():
    # 1. Setup the data
    labels = ['Performance Matrix', 'Stability', 'Time taken to respond', 'Accuracy', 'Complexity']
    stats = [85, 88, 90, 86, 70]

    # 2. Setup the Radar Chart structure
    num_vars = len(labels)
    
    # Compute angle for each axis
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
    
    # The plot is a circle, so we need to "close the loop"
    # We append the start value to the end of the list
    stats += stats[:1]
    angles += angles[:1]

    # 3. Create the Plot
    fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

    # Draw one axe per variable + add labels
    # We use angles[:-1] to avoid repeating the first label
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels, size=11, weight='bold')

    # Draw ylabels (Grid circles)
    ax.set_rlabel_position(0)
    plt.yticks([20, 40, 60, 80, 100], ["20", "40", "60", "80", "100"], color="grey", size=10)
    plt.ylim(0, 100)

    # 4. Plot the data
    ax.plot(angles, stats, linewidth=2, linestyle='solid', color='#1f77b4')
    ax.fill(angles, stats, '#1f77b4', alpha=0.25)

    # 5. Add Title
    plt.title('Gemma 2 (Med-Gemma)', size=20, weight='bold', y=1.1)

    # Show the plot
    plt.show()

if __name__ == "__main__":
    simulate_performance_graph()
```
